```python
import pandas as pd
```

```python
# reading the csv file
df_population=pd.read_csv('population_total.csv',encoding='latin1')
df_population
```

# 1 Making Pivot Table

## .pivot():Return reshaped DataFrame organized by given / column values('pivot without aggregation')

```python
# dropping null values 

df_population.dropna(inplace=True)
df_population
```

```python
# making a pivot table 
df_population_new=df_population.pivot(index='year',
                   columns='country',
                   values='population')
df_population_new
```

```python
# selecting some countries
df_population_new=df_population_new[['India','China','Brazil','Afghanistan']]
```

```python
# showing pivot table
df_population_new
```

```python
import matplotlib.pyplot as plt
```

# 2 Lineplot

```python
df_population_new.plot(kind='line')

plt.savefig('my_test.png') # if u want this plot png  save in your pc

plt.show() # extra information hide 
```

# 3 Barplot

## 3.1 Single Barplot 

```python
# selecting only one year (2020)
df_2020=df_population_new[df_population_new.index.isin([2020])]
```

```python
# showing new DataFrame
df_2020
```

```python
# transport DataFrame from row to columns or vicevera
df_2020=df_2020.T
```

```python
# making barplot
df_2020.plot(kind='bar',xlabel='country_name',
             ylabel='population',title='2020 pop. graph',
             color='pink')
```

## 3.2 Barplot grouped by 'n' variables


```python
df_five= df_population_new[df_population_new.index.isin([2000,2005,2010,2015,2020])]
```

```python
df_five.plot(kind='bar',xlabel='year',ylabel='population')
```
